1. What is the distribution of accident severity levels?

```python
print(ds['Severity'].value_counts().sort_index())
```

2. How is the accident frequency distributed by hour of the day?

```python
ds['Start_Time']=pd.to_datetime(ds['Start_Time'],errors='coerce')
ds['hour']=ds['Start_Time'].dt.hour
print(ds['hour'].value_counts().sort_index())
```

3. Which days of the week have the highest number of accidents?

```python
ds['day']=ds['Start_Time'].dt.day_name()
no=ds['day'].value_counts().sort_values(ascending=False)
print(no)
```

4. What are the most common weather conditions during accidents?

```python
weather=ds['Weather_Condition'].value_counts().sort_values(ascending=False)
print(weather.head())
```

5. How to identify columns with missing data and their missing percentage?

```python
data=ds.isnull().sum()
data=data[data!=0]

miss=pd.DataFrame({
    'count':data,
    'percentage':data/ds.shape[0]*100
})
print(miss)
```

6. How to impute missing numerical values with median?

```python
for col in ds.columns:
    if ds[col].dtype in ['int64', 'float64']:
        ds[col] = ds[col].fillna(ds[col].median())
print(ds.isnull().sum())
```

7. How to impute missing categorical values with the mode?

```python
for col in ds.columns:
    if ds[col].dtype == 'object':
        ds[col] = ds[col].fillna(ds[col].mode()[0])
print(ds.isnull().sum())
```

8. How to label encode a categorical column?

```python
ds.loc[:, 'City_Encoded'] = ds['City'].astype('category').cat.codes
print(ds[['City', 'City_Encoded']].head())
```

9. How to one-hot encode a categorical column?

```python
ds['City_Encoded'] = ds['City'].astype('category').cat.codes

print(ds[['City', 'City_Encoded']].head())
```
