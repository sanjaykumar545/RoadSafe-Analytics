```python
import pandas as pd
```

```python
ds=pd.read_csv("US_Accidents_March23.csv")
```

```python
#Droping airport code
ds.drop(['Airport_Code'],axis=1,inplace=True)
```

```python
#Handling missing data in precipitation and windspeed
temp=ds[['Wind_Speed(mph)', 'Precipitation(in)']][:100000]
print(temp.head())
```

```python

import pandas as pd
from sklearn.impute import KNNImputer
# Suppose temp_df is your large DataFrame with 7.7 million rows

# Step 1: Randomly sample 10,000 rows from the dataset
temp_sample=temp.sample(n=10000, random_state=42)

# Step 2: Perform KNN imputation on the sample
knnimputer=KNNImputer(n_neighbour=2)
sample=knnimputer.fit_transform(temp)
sample_ds=pd.DataFrame(sample,columns=temp_sample.columns)

#display
print(sample_ds.head())
```
