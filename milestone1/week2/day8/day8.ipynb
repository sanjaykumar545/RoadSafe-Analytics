```python
#Droping airport code
ds.drop(['Airport_Code'],axis=1,inplace=True)
```

```python
#Handling missing data in precipitation and windspeed
temp=ds[['Wind_Speed(mph)', 'Precipitation(in)']][:100000]
print(temp.head())
```

```python

from sklearn.impute import KNNImputer
# Suppose temp_df is your large DataFrame with 7.7 million rows

# Step 1: Randomly sample 10,000 rows from the dataset
temp_sample=temp.sample(n=10000, random_state=42)

# Step 2: Perform KNN imputation on the sample
knnimputer=KNNImputer(n_neighbors=2)
sample=knnimputer.fit_transform(temp)
sample_ds=pd.DataFrame(sample,columns=temp_sample.columns)

#display
print(sample_ds.head())
```

TASK: Impute the columns of 'Wind_Speed(mph)' & 'Weather_Conditions'

```python
from sklearn.impute import KNNImputer


temp_ds = ds[['Wind_Speed(mph)', 'Weather_Condition']][:100000]


knnimputer = KNNImputer(n_neighbors=2)
imputed = knnimputer.fit_transform(temp_ds[['Wind_Speed(mph)']])


temp_ds['Wind_Speed(mph)'] = imputed[:, 0]
ds.loc[:99999, 'Wind_Speed(mph)'] = imputed[:, 0]


temp_ds['Weather_Condition'] = temp_ds['Weather_Condition'].fillna(temp_ds['Weather_Condition'].mode()[0])


print(temp_ds[['Wind_Speed(mph)', 'Weather_Condition']].head())
```
