```python
missing=ds.isnull().sum()
missingper=missing/len(ds)*100
miss_summary=pd.DataFrame({
    'Miss Count':missing,
    'Miss Percentage':missingper
})
miss_summary=miss_summary[miss_summary['Miss Percentage']>=5]
print(miss_summary.sort_values(by='Miss Count',ascending=True))
```
