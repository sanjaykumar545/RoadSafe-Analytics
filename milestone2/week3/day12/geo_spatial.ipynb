```python
import pandas as pd
ds=pd.read_csv("/content/drive/My Drive/Infosys Internship/US_Accidents_March23.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
ds.dtypes
```

```python
ds[['Start_Lat','Start_Lng']].info()
```

```python
#dropping null values and duplicates
ds_geo=ds.dropna(subset=['Start_Lat','Start_Lng'])
ds_geo=ds_geo.drop_duplicates(subset=['Start_Lat','Start_Lng'])
```

```python
plt.figure(figsize=(10,8))
plt.scatter(ds_geo['Start_Lng'],ds_geo['Start_Lat'],alpha=0.1,s=1)
plt.title("US accidents hotspots(location)")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.kdeplot(
    x=ds_geo['Start_Lng'],
    y=ds_geo['Start_Lat'],
    cmap="Reds",fill=True,
    bw_adjust=0.5
)
plt.title("US accident density map")
plt.xlabel("Longitude")
plt.ylabel('Latitude')
plt.show()
```
